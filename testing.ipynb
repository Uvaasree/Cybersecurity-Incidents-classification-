```python
df=pd.read_csv('/content/drive/MyDrive/GUIDE_Test.csv')
```

```python
df.isnull().sum()
```

```python
#droping columns having null above 40 percent
forty_percent= (40*len(df))/100
for columns in df.columns:
    if df[columns].isnull().sum() > forty_percent:
        df.drop(columns=[columns],inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df[[
    'OrgId', 'IncidentId', 'DetectorId', 'AlertId', 'AlertTitle',
    'Category','IncidentGrade']]
ef=df.drop(columns=['IncidentGrade'])

from sklearn.preprocessing import LabelEncoder
# Label Encoding
label_encoder = LabelEncoder()

for i in ef.columns:
  df[i]=label_encoder.fit_transform(df[i])
```

```python
test_data=df[[
    'OrgId', 'IncidentId', 'DetectorId', 'AlertId', 'AlertTitle',
    'Category']]
```

```python
import pickle

with open('/content/drive/MyDrive/eq1_model.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
pred=model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

actual=df['IncidentGrade']
# Accuracy score
accuracy = accuracy_score(actual, pred)
print(f"Accuracy: {accuracy:.2f}")

# Classification report
print(classification_report(actual,pred))

# Confusion matrix
print(confusion_matrix(actual,pred))
```
